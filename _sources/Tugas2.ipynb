```python
# saya disuruh mebuat matrik A (3x3) B (3x3) C (3x2)
#

import numpy as np
A=np.array([[8,9,1],[1,2,3],[2,1,1]])
print("a\n", A)

B=np.array([[2,1,9],[4,3,2],[3,4,4]])
print("b\n", B)

C=np.array([[3,3],[1,2],[2,1]])
print("c\n", C)

D=np.array([[1,1,5],[4,3,2],[3,2,4]])
print("d\n", D)
```

```python
import numpy as np
A=np.array([[8,9,1],[1,2,3],[2,1,1]])
print("a\n", A)
print()

B=np.array([[2,1,9],[4,3,2],[3,4,4]])
print("b\n", B)
print()
print( "E=\n", A+B )

print("F=\n", 2*A)
print("G=\n", A@C)
```

```python
print(A@(B@C))

print((A@B)@C)
print("-"*30)
print(7*(A@B))
print((7*A)@B)
print(A@(7*B))
print("-"*30)
```

```python
C= np.array([[3,3],[1,2],[2,1]])
print("c\n", C)
C_T= C.transpose()
print(C_T)
print("-"*30)
print(C_T@C)
print("-"*30)
print(C@C_T)
```
